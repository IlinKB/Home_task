```python
import numpy as np
def kmeans(arr, k, max_iter, threshold):
    centre = arr[np.random.choice(len(arr) , k)]
    n = len(arr)
    m = len(arr[0])
    d = threshold  + 1
    k_iter = 0
    min_dist = np.zeros(n)
    while d > threshold and k_iter < max_iter:
        k_iter += 1
        new_centre = np.zeros((k, m))
        for i in range (0, n):
            min_dist[i] = np.argmin(np.linalg.norm(centre - arr[i], axis=1))
        for i in range(0, k):
            cluster_points = arr[min_dist == i]
            if len(cluster_points) > 0:
                new_centre[i] = np.mean(cluster_points, axis=0)
            else:
                new_centre[i] = centre[i]
        d = np.max(np.linalg.norm(centre - new_centre, axis=1))
        print(d)
        centre = new_centre.copy()
    print("суммарное квадратичное отклонение : " + str(d))
    print(min_dist)
    return centre
a = np.array([[1,2], [2,3], [1,1], [2,2]])
print(kmeans(a, 2, 1000, 10**-2))
```
