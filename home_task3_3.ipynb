```python
from sklearn.datasets import fetch_openml
import numpy as np
import matplotlib.pyplot as plt
#Загрузка может занять около минуты
images, labels = fetch_openml('mnist_784', return_X_y=True, as_frame=False)
i = 126
plt.imshow(images[i].reshape(28, 28), cmap='Greys_r')
plt.title('Number: ' + labels[i])
plt.show()
```

```python
centre = images[np.random.choice(len(images), 10, replace=False)]

min_dist = np.zeros(len(images))
d = 100
print(d)
while d > 10 ** -2:
    new_centre = np.zeros((10, 28 * 28))
    for k in range (0, len(images)):
        min_dist[k] = np.argmin(np.linalg.norm(centre - images[k], axis=1))
    for i in range(0, 10):
        cluster_points = images[min_dist == i]
        if len(cluster_points) > 0:
            new_centre[i] = np.mean(cluster_points, axis=0)
        else:
            new_centre[i] = centre[i]
    d = np.max(np.linalg.norm(centre - new_centre, axis=1))
    #print(d)
    centre = new_centre.copy()
    
            
            
        
        
```

```python
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
for i in range(10):
    ax = axes[i//5, i%5]
    ax.imshow(centre[i].reshape(28, 28), cmap='gray')
    ax.set_title(f'Images {i}')
    ax.axis('off')
plt.show()
```
