```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

data = np.loadtxt('https://lectures.scientific-python.org/_downloads/087d8fe936fcb7a4762b9850b932cf9e/populations.txt')

years = data[:, 0].astype(int)
hares = data[:, 1]
lynxes = data[:, 2]
carrots = data[:, 3]
plt.plot(years, hares, label='', marker='o')
plt.plot(years, lynxes, label='', marker='o')
plt.plot(years, carrots, label='', marker='o')
plt.xlabel('')
plt.ylabel('')
plt.title('  ,   ')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
max_hare_year = years[np.argmax(hares)]
max_lynx_year = years[np.argmax(lynxes)]
max_carrot_year = years[np.argmax(carrots)]

print(f"    : {max_hare_year}")
print(f"    : {max_lynx_year}")
print(f"    : {max_carrot_year}")


plt.plot(years, hares, label='', marker='o')
plt.plot(years, lynxes, label='', marker='o')
plt.plot(years, carrots, label='', marker='o')

plt.scatter(max_hare_year, np.max(hares), color='blue', s=100, zorder=5)
plt.scatter(max_lynx_year, np.max(lynxes), color='orange', s=100, zorder=5)
plt.scatter(max_carrot_year, np.max(carrots), color='green', s=100, zorder=5)
plt.xlabel('')
plt.ylabel('')
plt.title('     ')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
hare_lynx_total = hares + lynxes
hare_percentage = (hares / hare_lynx_total) * 100
lynx_percentage = (lynxes / hare_lynx_total) * 100


plt.bar(years, hare_percentage, label='', alpha=0.7)
plt.bar(years, lynx_percentage, bottom=hare_percentage, label='', alpha=0.7)
plt.xlabel('')
plt.ylabel('  (%)')
plt.title('      ')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].hist(hares, bins=8, alpha=0.7, color='blue', edgecolor='black')
axes[0].axvline(np.mean(hares), color='red', linestyle='--', linewidth=2, label=f': {np.mean(hares):.1f}')
axes[0].set_title('')
axes[0].set_xlabel('')
axes[0].set_ylabel('')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].hist(lynxes, bins=8, alpha=0.7, color='orange', edgecolor='black')
axes[1].axvline(np.mean(lynxes), color='red', linestyle='--', linewidth=2, label=f': {np.mean(lynxes):.1f}')
axes[1].set_title('')
axes[1].set_xlabel('')
axes[1].set_ylabel('')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

axes[2].hist(carrots, bins=8, alpha=0.7, color='green', edgecolor='black')
axes[2].axvline(np.mean(carrots), color='red', linestyle='--', linewidth=2, label=f': {np.mean(carrots):.1f}')
axes[2].set_title('')
axes[2].set_xlabel('')
axes[2].set_ylabel('')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"    :")
print(f": {np.mean(hares):.1f}  {np.std(hares):.1f}")
print(f": {np.mean(lynxes):.1f}  {np.std(lynxes):.1f}")
print(f": {np.mean(carrots):.1f}  {np.std(carrots):.1f}")
```

```python
dominant_species = np.where(hares > lynxes, 'H', 'L')
print("   :", dominant_species)

h_count = np.sum(dominant_species == 'H')
l_count = np.sum(dominant_species == 'L')

plt.pie([h_count, l_count], labels=[' (H)', ' (L)'], autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('      ')
plt.show()
```

```python
plt.plot(years, hares, label='', marker='o')
plt.plot(years, lynxes, label='', marker='o')
plt.xlabel('')
plt.ylabel('')
plt.title('    ')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

correlation = np.corrcoef(hares, lynxes)[0, 1]
print(f"      : {correlation:.3f}")
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(hares, lynxes, c=years, cmap='viridis', s=80, alpha=0.7)
plt.colorbar(scatter, label='')
plt.xlabel(' ')
plt.ylabel(' ')
plt.title('      ')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
years_without_first = years[1:]
hares_without_first = hares[1:]

slope, intercept, r_value, p_value, std_err = stats.linregress(years_without_first, hares_without_first)

predicted_1920 = intercept + slope * years[0]
actual_1920 = hares[0]
error_ratio = abs(predicted_1920 - actual_1920) / actual_1920

print(f"     1920 : {predicted_1920:.1f}")
print(f"     1920 : {actual_1920:.1f}")
print(f"    : {error_ratio:.3f} ({error_ratio*100:.1f}%)")
```
